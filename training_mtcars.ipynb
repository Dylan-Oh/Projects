## 이상치 확인 및 처리

```python
# 상그림 활용(이상치 : Q1, Q3로 부터 1.5*IQR을 초과하는 값)
```

```python
# 타이타닉 데이터 불러오기
df = sns.load_dataset('titanic')

# 상자그림
# 빅데이터분석기사 실기에서는 그림을 그리지 않는다.
sns.boxplot(df['age'])
```

```python
# Q1, Q3, IQR 구하기
Q1 = df['age'].quantile(.25)
Q3 = df['age'].quantile(.75)
IQR = Q3 - Q1 # IQR은 양수를 가져야 하기 때문에 큰수에서 작은수를 마이너스 해준다.
print(Q1, Q3, IQR)
print("---------------------")
print("Q1:", Q1, "Q3:", Q3, "IQR:", IQR)
```

```python
# 기준이 되는 limit 값을 구해준다.

upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
print(upper, lower) 

# age는 "나이"라는 변수의 특성상 양수만 존재하나, 그래도 최저값은 확인한다.
```

## 표준정규분포 활용(이상치 " +-3Z 값을 넘어가는 값)

```python
# 데이터 표준화, Z = (개별값-평균)/표준편차
# Z score로 변환
```

```python
mean_age = df['age'].mean() # 평균
std_age = df['age'].std() # 표준편차

print(mean_age)
print(std_age)
print("mean_age:", mean_age, "std_age:", std_age)
```

```python
# Z score 구하기

znorm = (df['age']-mean_age)/std_age
print(znorm)
# 중간에 결측치도 존재
```

```python
# 문제 : 이상치의 개수는 몇 개인가?(:+-3Z 기준)
# Z score 기준 3초과, 3미안인 값
```

```python
# Z score 3 초과인 것의 갯수 구하기

cond1 = (znorm > 3)
len(df[cond1])
```

```python
# Z score 3 미만인 것의 갯수 구하기

cond2 = (znorm < -3)
len(df[cond2])
```

```python
# Z score 3초과, 3미만인 값의 총계

print(len(df[cond1]) + len(df[cond2]))
```

## 데이터 합치기

```python
# 행, 열 방향으로 데이터 합치기

df = sns.load_dataset('iris') # iris 데이터셋 불러오기
df.head()
```

```python
df.shape
```

### 데이터를 행방향으로 결합

```python
# 데이터 2개로 분리, loc
# df1, df2로 행을 나누어 각각의 변수에 데이터 저장

df1 = df.loc[0:30, ]
df2 = df.loc[31:60, ]
```

```python
df1.head()
```

```python
df2.head()

# df1과 시작하는 인덱싱이 다름을 통해 데이터셋을 정상적으로 잘 분리된 것을 확인
```

```python
# axis=0 행
# axis=1 열
# 판다스 concat 함수 사용

df_sum = pd.concat([df1, df2], axis=0) # 행 방향으로 결합(위, 아래)
print(df_sum.head())
print(df_sum.shape)
```

### 데이터를 열방향으로 결합

```python
# 데이터 2개로 나누기

df1 = df.loc[: , 'sepal_length':'petal_length'] # 1~3열 추출 데이터
df2 = df.loc[: , ['petal_width', 'species']] # 4~5열 데이터
```

```python
# axis=1 열

df_sum = pd.concat([df1, df2], axis=1) # 열방향으로 결합(좌, 우)
df_sum.head()
```

## 날짜/시간 데이터, index 다루기

### 날짜 다루기

```python
# 데이터 만들기

df = pd.DataFrame({
    '날짜' : ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230521'],
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A' ],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600]
}
)
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# 데이터 타입 datetime 으로 변경
# datetime64[ns]로 변경 된 것을 확인할 수 있다.

df['날짜'] = pd.to_datetime(df['날짜'])
df.info()
```

```python
# 날짜 데이터에서 년, 월, 일 변수(열) 추가하기

df['year'] = df['날짜'].dt.year
df['month'] = df['날짜'].dt.month
df['day'] = df['날짜'].dt.day
df
```

```python
df.info()
```

```python
# 날짜 구간 필터링
# between

df [df['날짜'].between('2023-01-01', '2023-01-31')] # 1월31일은 미포함, between 함수에서는 뒤의 것을 포함하지 않는다.
```

```python
# 날짜를 인덱스로 설정 후 loc 함수 사용
# 데이터 만들기

df = pd.DataFrame({
    '날짜' : ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230521'],
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A' ],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600]})

# 데이터 타입 datetime으로 변경(필수)
df['날짜'] = pd.to_datetime(df['날짜'])
df = df.set_index('날짜') # drop=True(디폴트) or False
# df = df.set_index('날짜', drop=False) # 기존의 날짜 변수를 삭제하지 않고, 그대로 비교해서 보여줄 필요가 있을 경우, drop=False로 옵션 설정
df.head(3)
```

```python
# 데이터 타입을 datetime으로 변경한 이후 loc 함수 사용하여 필터링

print(df.loc['2023-01-05' : '2023-02-23']) #둘 다 기간 포함
print(df.loc[(df.index>='2023-01-05') & (df.index<='2023-02-23')])
```

### 시간 다루기

```python
# 시간 데이터 만들기

df = pd.DataFrame({
    '물품' : ['A', 'B', 'A', 'B', 'A', 'B', 'A' ],
    '판매수' : [5, 10, 15, 15, 20, 25, 40],
    '개당수익' : [500, 600, 500, 600, 600, 700, 600]})

time = pd.date_range('2023-09-24 12:25:00', '2023-09-25 14:45:00', periods=7) # periods는 몇 개로 나눌지 설정
df['time']=time
df = df[['time', '물품', '판매수', '개당수익']]
df
```

```python
print(df)
```

```python
# index 초기화(인덱스를 컬럼으로)
# df = df.reset_index()
# df
# 기존에 있던 인덱스에 인덱스 명을 컬럼으로 새로 추가할 때
```

```python
# index 새로 지정하는 경우

df = df.set_index('time')
df
# 기존에 인덱스로 표출하던 "0 ~ " 숫자는 표출하지 않는다. 인덱스를 새로 지정하였기 때문에
```

```python
# 시간 데이터 다루기(주의 : 시간이 index에 위치해야 함)
# between_time 함수 사용
# 시간만 필터링을 하기 때문에, 다른 날짜의 시간도 포함하여 표출한다.

df.between_time(start_time='12:25', end_time='21:00') # 시작 시간, 끝 모두 포함

# include_start=False, include_end=False 옵션으로 시작, 끝 시간 제외 가능
```

```python
# 날짜를 인덱스로 설정한 이후 loc 함수 사용

print(df.loc['2023-09-24 12:25:00':'2023-09-24 21:11:50']) # 둘 다 포함
print(df.loc[(df.index>='2023-09-24 12:25:00') & (df.index<='2023-09-24 21:11:50')])
```
